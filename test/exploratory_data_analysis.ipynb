PCA Analysis 

```python
import pandas as pd 
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
```

```python
data = pd.read_csv('filtered_sorted_dataset.csv') 
data.head()
```

```python
# Exclude the document names column from the DataFrame
new_data = data.drop(columns=['Document']) 
new_data
```

```python
# Assume 'data' is your cleaned and standardized dataset with 56 numerical features
# Standardize the data
scaler = StandardScaler()
scaled_data = scaler.fit_transform(new_data)

# Initialize PCA with desired number of components
pca = PCA(n_components=55)  # You can adjust the number of components based on your analysis needs

# Fit PCA to the scaled data and transform it
pca_data = pca.fit_transform(scaled_data)

# Analyze principal components, explained variance, and loadings
explained_variance_ratio = pca.explained_variance_ratio_
loadings = pca.components_

# Visualize the results if needed
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Assume 'loadings' contains the loadings of each principal component
# Convert loadings to a DataFrame for easier plotting
df_loadings = pd.DataFrame(loadings.T, columns=[f'PC{i+1}' for i in range(loadings.shape[0])], index=new_data.columns)

# Create a heatmap
plt.figure(figsize=(12, 8))
sns.heatmap(df_loadings, cmap='coolwarm', annot=True, fmt=".2f")
plt.title('Loadings of Principal Components')
plt.xlabel('Principal Components')
plt.ylabel('Features')
plt.show()
```

```python
# Identify columns with 1, 2, or 3 unique values
columns_to_drop = []
for column in new_data.columns:
    unique_values = new_data[column].nunique()
    if unique_values <= 10:  # Change this threshold as needed
        columns_to_drop.append(column)

# Drop identified columns
new_data.drop(columns=columns_to_drop, inplace=True) 

new_data.head
```
